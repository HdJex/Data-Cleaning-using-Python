# outlier detection

```python
df.head(2)
```

```python
Q1 = df['number_of_reviews'].quantile(0.25)
Q3 = df['number_of_reviews'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['number_of_reviews'] < lower_bound) | (df['number_of_reviews'] > upper_bound)]
outliers.head()
```

```python
data = df[(df['number_of_reviews'] > lower_bound) & (df['number_of_reviews'] < upper_bound)]
data.describe()
```
